```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters (set these!)
y0 = 1.0        # initial displacement
yprime0 = 0.0   # initial velocity
z = 0.1         # damping ratio
w0 = 2.0        # natural frequency (rad/s)

# Derived quantity
wd = w0*np.sqrt(1 - z**2)  # damped natural frequency

# Constants
A = y0
B = (yprime0 + z*w0*y0) / wd

# Time axis
t = np.linspace(0, 20, 400)

# Solution form
y = np.exp(-z*w0*t) * (A*np.cos(wd*t) + B*np.sin(wd*t))

# Plot
plt.plot(t, y)
plt.title("Damped Oscillator")
plt.xlabel("t")
plt.ylabel("y(t)")
plt.grid(True)
plt.show()
```
